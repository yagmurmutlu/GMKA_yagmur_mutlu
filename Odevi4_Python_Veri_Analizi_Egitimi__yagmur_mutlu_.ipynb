### Veriyi özetleme

1.  3. Ders sonunda hazırladığınız 2021 ve 2022 yıllarına ait BIST30 ve EUR/USD kur değerlerini içeren bileşke tablodan sadece ayın son işgününe ait değerleri içeren bir pivot tablo üretiniz.

```python
import yfinance as yf
import plotly.express as px
import pandas as pd
import matplotlib.pyplot as plt
import plotly.graph_objects as go
```

```python
indirme_listesi= ['XU030.IS','EURTRY=X']

veri_2022= pd.DataFrame()

for i in indirme_listesi:
  data=yf.download(i,start='2022-01-01', end = '2022-12-31')
  data=data['Adj Close'].rename(i)
  veri_2022 = pd.concat([veri_2022,data],axis=1)

veri_2022.columns = ['Kapanış (TRY)','EUR/TRY']
veri_2022 = veri_2022[veri_2022['Kapanış (TRY)']>0]
norm = veri_2022.iloc[0,0]
norm2 = veri_2022.iloc[0,1]
veri_2022['Normalize'] = veri_2022['Kapanış (TRY)']/norm
veri_2022['Normalize EUR/TRY'] = veri_2022['EUR/TRY']/norm2
veri_2022['degisim'] = veri_2022['Normalize'].diff(1)
veri_2022['degisim%'] =( veri_2022['degisim']/veri_2022['Normalize'].shift(1))*100
veri_2022['d_yon'] = veri_2022['degisim'].map(lambda x : 'dusen gun' if x < 0 else'yukselen gun')
veri_2022
```

```python
indirme_listesi= ['XU030.IS','EURTRY=X']
veri_2021= pd.DataFrame()

for i in indirme_listesi:
  data=yf.download(i,start='2021-01-01', end = '2021-12-31')
  data=data['Adj Close'].rename(i)
  veri_2021 = pd.concat([veri_2021,data],axis=1)
  

veri_2021.columns = ['Kapanış (TRY)','EUR/TRY']
veri_2021 = veri_2021[veri_2021['Kapanış (TRY)']>0]
norm = veri_2021.iloc[0,0]
veri_2021.iloc[0,0]
norm2 = veri_2021.iloc[0,1]
veri_2021['Normalize'] = veri_2021['Kapanış (TRY)']/norm
veri_2021['Normalize EUR/TRY'] = veri_2021['EUR/TRY']/norm2
veri_2021['degisim'] = veri_2021['Normalize'].diff(1)
veri_2021['degisim%'] =( veri_2021['degisim']/veri_2021['Normalize'].shift(1))*100
veri_2021['d_yon'] = veri_2021['degisim'].map(lambda x : 'dusen gun' if x < 0 else'yukselen gun')
veri_2021
```

```python
veri = pd.concat([veri_2021,veri_2022],axis=0)
veri
```

```python
veri['ay_yıl']=veri.index.strftime('%Y-%m')
veri
```

```python
veri_as=veri.pivot_table(values=['Kapanış (TRY)	'],index='ay_yıl',aggfunc='last')
veri
```

2. Ay kırılımlı olarak ürettiğiniz tabloyu grafik olarak inceleyiniz.


```python
veri_as['degisim'] = veri_as['Kapanış(TRY)'].diff(1)
veri_as
```

```python
veri_as['degisim'] = veri_as['Kapanış(TRY)'].diff(1)
veri_as['d_yon']=veri_as['degisim'].map(lambda x:'düşen gün' if x<0 else 'yükselen gün')
veri_as['degisim%']=(veri_as['degisim']/veri_as['Kapanış(TRY)'].shift(1))
veri_as
```

```python
px.bar(veri_as,y='degisim%',
       color ='d_yon',color_discrete_map={'düşen gün':'red','yükselen gün':'green'},
       text_auto='.0%'
       )
```

3. Ders sonunda hazırladığınız 2021 ve 2022 yıllarına ait BIST30 ve EUR/USD kur değerlerini içeren bileşke tablodan sadece ayın son işgününe ait değerleri içeren bir tabloyu bir de resample yöntemi ile üretiniz.

```python
px.bar(veri_as2,y='degisim%',
       color ='d_yon',color_discrete_map={'düşen gün':'red','yükselen gün':'green'},
       text_auto='.0%'
       )
```
