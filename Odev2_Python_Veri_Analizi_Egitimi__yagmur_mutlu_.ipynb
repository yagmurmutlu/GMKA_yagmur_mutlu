**DERS 2 **

VERİ TEMİZLEME

```python
#SAtır silme

B30 =B30 [B30.index> '2021-01-01']
B30
```

Bölüm 3:
 Veri temizleme:
• Tabloda sadece "Adj Close" (düzeltilmiş kapanış) sütununu bırakınız.
• "Adj Close" sütununu " Kapanış (TRY)" olarak yeniden adlandırınız.

```python
#sütün silme

B30 = B30[['Adj Close']] #kapanış değeri

#B30= B30[['Open','Adj Close']] #açılış değerini siler
B30
```

```python
#isim değiştirme

B30.columns = ['Kapanış (TRY)']
B30
```

Bölüm 4:

Tabloya yeni sütun ekleme:
Tabloya yeni bir sütun ekleyerek Kapanış (TRY) değerlerini 1 Ocak 2021 tarihindeki kapanış değerine göre normalize ediniz (yani tüm değerleri 1 Ocak 2021 değerine bölünüz). 

```python
norm = B30.iloc[0,0]
norm

B30['Normalize'] = B30 ['Kapanış (TRY)']/norm
B30
```

**BIST 30 endeksi 2021 yılı başında sonuna kadar yaklaşık 1.24 kadar çıktı**

Bölüm 5:

BIST30 endeksinin 2021 yılı boyunca değişimini yorumlayınız.

```python
B30['Normalize'].plot()
```

```python
fig1= px.scatter(B30,  y='Normalize')
fig2=px.bar(B30,  y='Normalize')
fig3=px.line(B30,  y='Normalize')
fig1.show()
fig2.show()
fig3.show()
```

Bölüm 6:


Değişim alanlarını ekleme ve grafikte koşullu renklendirme:
• Normalize etiğiniz alan için her satırda bir önceki satırdaki Normalize değeri ile olan farkını hesaplattırınız (bu sütunu "degisim" olarak isimlendirebilirsiniz). • Tabloyu px.bar grafiği olarak görselleştiriniz.
• px.bar grafiğinde endeksin yükseldiği günleri "yeşil", düştüğü günleri "kırmızı" olarak görselleştiriniz.

```python
fig_1 = px.line(B30,  y='Normalize',
        width=1000 , height =500,
        title = 'BİST30 Endeksi 2021- Normalize (1 ocak 2021)',
        hover_data={'Normalize': ':.2f' , 'Kapanış (TRY)' : ': 0f'},
        labels = {'Date' : 'Tarih' , 'Normalize' : 'BIST30 Endeksi (TRY)'})

fig_1.show()
```

**DERS 4**
Koşullu GRAFİK RENKLENDİRME

```python
B30['değişim'] = B30['Normalize'].diff(1)
```

```python
B30['d_yon'] =B30['değişim'].map(lambda x: 'düşen gün' if  x <0 else 'yükselen gün')
B30
```

```python
B30['değişim%'] = (B30['değişim']/B30['Normalize'].shift(1))*100
B30
```

```python
px.bar(B30, y= 'değişim%',
       color = 'd_yon', color_discrete_map={'düşen gün': 'red', 'yükselen gün' : 'green'})
```

```python
fig_3= px.histogram(B30['değişim'], nbins =10)
fig_3.show()
```
